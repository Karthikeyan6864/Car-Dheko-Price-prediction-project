```python
#final df for chennai_cars all collum parsed
import pandas as pd
import ast


file_path = 'chennai_cars.xlsx'
data = pd.read_excel(file_path)


def parse_new_car_detail(details_data):
    if isinstance(details_data, str):
        return ast.literal_eval(details_data)
    return {}


def parse_car_overview(overview_data):
    overview_dict = {}
    if isinstance(overview_data, dict) and 'top' in overview_data:
        for item in overview_data['top']:
            if 'key' in item and 'value' in item:
                overview_dict[item['key']] = item['value']
    return overview_dict


def parse_car_features(features_data):
    features_dict = {}
    if isinstance(features_data, dict):
        if 'top' in features_data:
            for item in features_data['top']:
                if 'value' in item:
                    features_dict[item['value']] = True  # Mark each feature as True if present
        if 'data' in features_data:
            for section in features_data['data']:
                heading = section.get('heading', 'Unknown Heading')
                sub_heading = section.get('subHeading', 'Unknown Subheading')
                for item in section.get('list', []):
                    if 'value' in item:
                        key = f"{heading} - {sub_heading} - {item['value']}"
                        features_dict[key] = True
    return features_dict


def parse_car_specs(specs_data):
    specs_dict = {}
    if isinstance(specs_data, dict):
        if 'top' in specs_data:
            for item in specs_data['top']:
                if 'key' in item and 'value' in item:
                    specs_dict[item['key']] = item['value']
        if 'data' in specs_data:
            for section in specs_data['data']:
                heading = section.get('heading', 'Unknown Heading')
                sub_heading = section.get('subHeading', 'Unknown Subheading')
                for item in section.get('list', []):
                    if 'key' in item and 'value' in item:
                        key = f"{heading} - {sub_heading} - {item['key']}"
                        specs_dict[key] = item['value']
    return specs_dict

# Apply parsing functions to respective columns
details_df = data['new_car_detail'].apply(lambda x: parse_new_car_detail(x))
overview_df = data['new_car_overview'].apply(lambda x: parse_car_overview(ast.literal_eval(x) if isinstance(x, str) else {}))
feature_df = data['new_car_feature'].apply(lambda x: parse_car_features(ast.literal_eval(x) if isinstance(x, str) else {}))
specs_df = data['new_car_specs'].apply(lambda x: parse_car_specs(ast.literal_eval(x) if isinstance(x, str) else {}))

# Convert the parsed data to DataFrames
details_df = pd.DataFrame(details_df.to_list())
overview_df = pd.DataFrame(overview_df.to_list())
feature_df = pd.DataFrame(feature_df.to_list())
specs_df = pd.DataFrame(specs_df.to_list())

# Combine all parsed DataFrames with the original data (excluding original columns parsed)
final_df = pd.concat([data.drop(columns=['new_car_detail', 'new_car_overview', 'new_car_feature', 'new_car_specs']),
                      details_df, overview_df, feature_df, specs_df], axis=1)

# Display the first few rows of the combined DataFrame
final_df.head()
```

```python
# Concatenate all DataFrames into one
import pandas as pd

dfs = [final_df, final_df1, final_df2, final_df3, final_df4, final_df5]

cities = ['chennai', 'bangalore', 'delhi', 'hyderabad', 'jaipur', 'kolkata']

# Add the 'city' column to each DataFrame
for df, city in zip(dfs, cities):
    df['city'] = city

common_columns = set(dfs[0].columns)  # Start with columns of the first DataFrame
for df in dfs[1:]:
    common_columns &= set(df.columns)  # Keep only the common columns

# Filter each DataFrame to include only the common columns
dfs_filtered = [df[list(common_columns)] for df in dfs]

merged_df = pd.concat(dfs_filtered, ignore_index=True)

print(merged_df.shape)
print(merged_df.head())
```

```python
# Calculate missing values for each column
import pandas as pd
from google.colab import files

missing_values = merged_df.isnull().sum()

missing_data = pd.DataFrame({
    'Column Name': missing_values.index,
    'Missing Values': missing_values.values
})

file_path = '/content/missing_values_per_column.xlsx'
missing_data.to_excel(file_path, index=False)

files.download(file_path)

```

```python
# # Drop the columns with more than 15% missing values
missing_percentage = merged_df.isnull().mean() * 100

columns_to_drop = missing_percentage[missing_percentage > 15].index

merged_df_cleaned = merged_df.drop(columns=columns_to_drop)

print(merged_df_cleaned.shape)
print(merged_df_cleaned.head())
```

```python
# Print the column names that contain '-'

column_names = merged_df_cleaned.columns

columns_with_hyphen = [col for col in column_names if '-' in col]

print("Columns with hyphen:", columns_with_hyphen)
```

```python
# Rename the columns: keep only the part after the last '-'
column_names = merged_df_cleaned.columns

new_column_names = [col.rsplit('-', 1)[-1] for col in column_names]

merged_df_cleaned.columns = new_column_names

print("Updated Column Names:", merged_df_cleaned.columns)
```

```python
# Convert 'TRUE'/'FALSE' strings to boolean values (True/False)
# cleaning Boolian values
import pandas as pd


object_columns = merged_df_cleaned.select_dtypes(include=['object']).columns
print(f"Object columns identified: {object_columns}")


print("\nFirst 5 rows of columns with 'TRUE'/'FALSE' as strings:")
print(merged_df_cleaned[object_columns].head())

boolean_columns = []
for col in object_columns:
    # Check if the column contains 'TRUE'/'FALSE' values (case-insensitive)
    if merged_df_cleaned[col].apply(lambda x: str(x).strip().lower() in ['true', 'false']).any():
        boolean_columns.append(col)
        print(f"Column '{col}' contains 'TRUE' or 'FALSE' values (as strings).")


for col in boolean_columns:

    merged_df_cleaned[col] = merged_df_cleaned[col].apply(
        lambda x: True if str(x).strip().lower() == 'true' else (False if str(x).strip().lower() == 'false' else x)
    )


for col in boolean_columns:
    merged_df_cleaned[col] = merged_df_cleaned[col].fillna(False)

print("\nFirst 5 rows after filling NaN with False in boolean columns:")
print(merged_df_cleaned[boolean_columns].head())

print("\nBoolean columns after conversion:")
print(merged_df_cleaned[boolean_columns].dtypes)
```

```python
# Clean up column names (strip spaces and standardize case if needed)
merged_df_cleaned.columns = merged_df_cleaned.columns.str.strip()  # Remove leading/trailing spaces

merged_df_cleaned = merged_df_cleaned.loc[:, ~merged_df_cleaned.columns.duplicated()]


print(merged_df_cleaned)
```

```python
columns_to_drop = ['Turbo Charger', 'RTO','priceActual','priceSaving','it','car_links','trendingText','variantName',]
df = merged_df_cleaned.drop(columns=columns_to_drop)
df_description = df.describe()
print(df_description)
```
