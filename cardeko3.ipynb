```python
import re

# Function to extract the first numeric part
def extract_first_numeric(value):
    if isinstance(value, str):  # Ensure the value is a string
        nums = re.findall(r'\d+', value)  # Find all numeric parts
        if nums:
            return int(nums[0])  # Return the first numeric part as an integer
    return None  # Return None if no numeric value is found

# List of columns to process
columns_to_clean = ['Kerb Weight', 'Height', 'Length', 'Width','Wheel Base','Seating Capacity','Displacement','No Door Numbers','Values per Cylinder','Engine','Mileage']  # Add more column names as needed

# Convert any non-numeric columns to strings
for col in columns_to_clean:
    df[col] = df[col].astype(str)

# Process each column
for col in columns_to_clean:
    new_col = f'{col} Cleaned'  # Create a new column name for cleaned data
    df[new_col] = df[col].apply(extract_first_numeric)  # Apply cleaning function

# Drop rows where any of the new cleaned columns have missing values
df_cleaned = df.dropna(subset=[f'{col} Cleaned' for col in columns_to_clean])

# Drop the original columns if no longer needed
df_cleaned = df_cleaned.drop(columns=columns_to_clean, errors='ignore')

# Rename cleaned columns back to the original names
df_cleaned.rename(columns={f'{col} Cleaned': col for col in columns_to_clean}, inplace=True)

# Check the cleaned DataFrame
print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")

```

```python
#cleaning price column
import pandas as pd



def convert_price(value):
    value = value.replace('₹', '').replace(',', '').strip()  # Remove ₹ and commas
    if 'crore' in value:
        return float(value.replace('crore', '').strip()) * 1e7
    elif 'lakh' in value:
        return float(value.replace('lakh', '').strip()) * 1e5
    else:
        return float(value)


df_cleaned['price'] = df_cleaned['price'].apply(convert_price)

print(df_cleaned)
```

```python
# finding unique values in columns
non_numeric_columns = df_cleaned.select_dtypes(exclude=['number'])

for col in non_numeric_columns.columns:
    non_unique_values = df_cleaned[col].value_counts()[df_cleaned[col].value_counts() > 1]
    if not non_unique_values.empty:
        print(f"Column: {col}")
        print(non_unique_values)
        print("-" * 40)
```

```python
# Define the columns to clean
categorical_columns_to_clean = ['Color', 'Rear Brake Type', 'Tyre Type', 'Insurance Validity',
                                 'Steering Type', 'Front Brake Type']

# Define mappings for each column to group similar values
mappings = {
    'Rear Brake Type': {
        'drum': ['drum', 'drums', 'drum`', 'drum in discs', 'self-adjusting drum', 'leading-trailing drum',
                 'leading & trailing drum', 'self adjusting drums', 'booster assisted drum',
                 'mechanical-hydraulic dual circuit', 'vacuum assisted hydraulic dual circuit w'],
        'disc': ['disc', 'discs', 'solid disc', 'ventilated disc', 'ventilated discs', 'ventialte disc',
                 'disc & caliper type', 'ventlated disc', 'ventillated disc', 'ventilated drum',
                 'disc & drum', 'multilateral disc', 'drum in disc', 'caliper ventilated disc', 'disk'],
    },
    'Tyre Type': {
        'tubeless radial': ['tubeless,radial', 'tubeless, radial', 'radial, tubeless', 'radial tubeless',
                            'tubeless radial tyres', 'tubeless radials tyre', 'tubeless tyres, radial'],
        'tubeless': ['tubeless', 'tubeless tyres', 'tubeless tyre', 'tubless, radial', 'tubeless,runflat',
                     'tubeless. runflat'],
        'runflat': ['run-flat', 'runflat', 'runflat tyres', 'runflat tyre'],
    },
    'Insurance Validity': {
        'third party': ['third party insurance', 'third party'],
        'comprehensive': ['comprehensive', 'zero dep', 'not available'],
    },
    'Steering Type': {
        'electric': ['electric', 'electrical', 'electronic', 'epas'],
        'power': ['power', 'hydraulic', 'manual'],
    },
    'Front Brake Type': {
        'disc': ['disc', 'disk', 'solid disc', 'ventilated disc', 'ventilated discs', 'ventilated disk',
                 'vantilated disc', 'ventillated disc', 'ventlated disc', 'disc, 236 mm',
                 'caliper ventilated disc', 'ventillated disc', 'booster assisted ventilated disc'],
        'drum': ['drum', 'booster assisted drum'],
    },
    'Color': {
        'white': ['white', 'polar white', 'superior white', 'diamond white', 'tafeta white', 'candy white'],
        'grey': ['grey', 'gray', 'titanium grey', 'granite grey', 'modern steel metallic', 'mist silver'],
        'black': ['black', 'phantom black', 'aurora black pearl'],
        'red': ['red', 'fiery red', 'berry red', 'maroon'],
        'silver': ['silver', 'silky silver', 'sleek silver'],
        'blue': ['blue', 'nexa blue', 'dark blue'],
        'brown': ['brown', 'golden brown', 'beige'],
        'gold': ['gold', 'golden'],
        'green': ['green', 'copper'],
    }
}

# Function to clean and categorize a column based on the mapping
def clean_column(column, mapping):
    if column in mappings:
        col_mapping = mappings[column]
        return df_cleaned[column].replace({
            value: key for key, values in col_mapping.items() for value in values
        })
    return df_cleaned[column]


for column in categorical_columns_to_clean:
    if column in mappings:
        df_cleaned[column] = clean_column(column, mappings)

for column in categorical_columns_to_clean:
    print(f"Unique values in '{column}' after cleaning:")
    print(df_cleaned[column].unique())
    print("-" * 50)
```

```python
import pandas as pd
import re



# Function to clean and categorize 'Color' column
def clean_color(color):
    color = color.lower()
    if re.search(r'(grey|gray|smoke|magma)', color):
        return 'grey'
    elif 'black' in color:
        return 'black'
    elif 'white' in color:
        return 'white'
    elif 'red' in color:
        return 'red'
    elif 'blue' in color:
        return 'blue'
    elif 'green' in color:
        return 'green'
    elif 'brown' in color:
        return 'brown'
    elif 'yellow' in color:
        return 'yellow'
    elif 'silver' in color:
        return 'silver'
    elif 'orange' in color:
        return 'orange'
    elif 'gold' in color:
        return 'gold'
    elif 'violet' in color:
        return 'violet'
    elif 'purple' in color:
        return 'purple'
    else:
        return 'miscellaneous'


def clean_rear_brake_type(brake_type):
    brake_type = brake_type.lower()
    if 'disc' in brake_type:
        return 'disc'
    elif 'drum' in brake_type:
        return 'drum'
    elif 'electric parking brake' in brake_type:
        return 'electric parking brake'
    else:
        return 'self adjusting drum'


def clean_tyre_type(tyre_type):
    tyre_type = tyre_type.lower()
    if 'tubeless' in tyre_type and 'radial' in tyre_type:
        return 'tubeless radial'
    elif 'tubeless' in tyre_type:
        return 'tubeless'
    elif 'radial' in tyre_type:
        return 'radial'
    elif 'runflat' in tyre_type:
        return 'runflat'
    elif 'mud terrain' in tyre_type:
        return 'mud terrain'
    elif 'all terrain' in tyre_type:
        return 'all terrain'
    else:
        return 'specialized'


def clean_insurance_validity(insurance):
    if insurance in ['third party', 'comprehensive']:
        return insurance
    elif insurance in ['1', '2']:
        return 'valid'
    else:
        return 'expired'


def clean_steering_type(steering_type):
    steering_type = steering_type.lower()
    if 'electric' in steering_type:
        return 'electric'
    elif 'power' in steering_type:
        return 'power'
    else:
        return 'manual'


def clean_front_brake_type(brake_type):
    brake_type = brake_type.lower()
    if 'disc' in brake_type:
        return 'disc'
    elif 'drum' in brake_type:
        return 'drum'
    elif 'abs' in brake_type:
        return 'abs'
    else:
        return 'vacuum assisted'



df_cleaned['Color'] = df_cleaned['Color'].apply(clean_color)
df_cleaned['Rear Brake Type'] = df_cleaned['Rear Brake Type'].apply(clean_rear_brake_type)
df_cleaned['Tyre Type'] = df_cleaned['Tyre Type'].apply(clean_tyre_type)
df_cleaned['Insurance Validity'] = df_cleaned['Insurance Validity'].apply(clean_insurance_validity)
df_cleaned['Steering Type'] = df_cleaned['Steering Type'].apply(clean_steering_type)
df_cleaned['Front Brake Type'] = df_cleaned['Front Brake Type'].apply(clean_front_brake_type)


print(df_cleaned)

```
