```python
import pandas as pd
df = pd.read_excel('cardeko (1).xlsx')
df
```

```python
df.drop(columns=['Ownership','ft','Kms Driven','Year of Manufacture','Registration Year','Max Torque'], inplace=True)
```

```python
#  Convert those string columns to lowercase
import pandas as pd

string_columns = df.select_dtypes(include=['object']).columns

df[string_columns] = df[string_columns].apply(lambda x: x.str.lower())

print(df.head())
```

```python
# Find missing values count for all columns (including those with no missing values)
import pandas as pd
from google.colab import files

missing_values = df.isnull().sum()

missing_values_df = missing_values.reset_index()

missing_values_df.columns = ['Column', 'Missing Values Count']

missing_values_df['Missing Values Count'] = missing_values_df['Missing Values Count'].fillna(0).astype(int)

missing_values_df.to_excel('missing_values_report.xlsx', index=False)

print(missing_values_df)

files.download('missing_values_report.xlsx')
```

```python
df.drop(columns=['Seats','Engine Displacement'], inplace=True)
```

```python
df['Max Power'] = df['Max Power'].astype(str).str.extract(r'(\d+\.\d+|\d+)').astype(float)
df['Torque'] = df['Torque'].astype(str).str.extract(r'(\d+\.\d+|\d+)').astype(float)

df[['Torque', 'Max Power']]
```

```python
#Handling missing values by mean and mode imputation
import pandas as pd
import numpy as np



# List of columns to apply mean imputation (numerical)
numerical_columns = ['Max Power', 'Torque']

# List of columns to apply mode imputation (categorical)
categorical_columns = ['Engine Type', 'Mileage', 'Steering Type', 'Wheel Base', 'Gear Box', 'Rear Brake Type',
                       'Front Brake Type', 'Width', 'Length', 'Height', 'Tyre Type', 'Kerb Weight',
                       'Values per Cylinder', 'No of Cylinder', 'No Door Numbers', 'Seating Capacity',
                       'bt', 'Displacement', 'Engine', 'Color']

# Insurance Validity (categorical column) - grouped by 'owner'
Insurance_validity = ['Insurance Validity']

# Fallback columns dictionary
for col in numerical_columns:
    df[col] = df.groupby('model')[col].transform(lambda x: x.fillna(x.mean()))

# Perform mode imputation for categorical columns based on 'model'
for col in categorical_columns:
    overall_mode = df[col].mode().iloc[0] if not df[col].mode().empty else None  # Overall mode of the column
    df[col] = df.groupby('model')[col].transform(
        lambda x: x.fillna(x.mode().iloc[0] if not x.mode().empty else overall_mode)
    )
# Perform mode imputation for 'Insurance Validity' based on 'owner'
for col in Insurance_validity:
    df[col] = df.groupby('owner')[col].transform(
        lambda x: x.fillna(x.mode()[0] if not x.mode().empty else None)
    )

# Check the dataset for remaining missing values
missing_values_after_imputation = df.isnull().sum()
missing_values_after_imputation
```

```python
# List to store columns with the specified values("unknown")
columns_with_unknown = []


for col in df.columns:

    count = df[col].isin(['Unknown', 'unknown', None]).sum()
    if count > 0:
        columns_with_unknown.append((col, count))


for col, count in columns_with_unknown:
    print(f"Column: {col}, Count: {count}")
```

```python
# Filter out rows where 'Max Power' or 'Torque' are blank or NaN
df_cleaned = df[~(df['Max Power'].isna() | df['Torque'].isna())]

print(f"Original DataFrame shape: {df.shape}")
print(f"Cleaned DataFrame shape: {df_cleaned.shape}")
```

```python
# Combine checks for blank values and NaN
columns_with_blanks_or_missing = []


for col in df_cleaned.columns:
    if df_cleaned[col].dtype == 'object':
        count = (df_cleaned[col].str.strip() == '').sum()
    else:
        count = df_cleaned[col].isna().sum()

    if count > 0:
        columns_with_blanks_or_missing.append((col, count))

for col, count in columns_with_blanks_or_missing:
    print(f"Column: {col}, Blank or Missing Value Count: {count}")
```
